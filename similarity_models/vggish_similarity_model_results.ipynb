# Song similarity

In this notebook, I will our vggish siilarity model to find the most similar songs of songs across multiple genres.

```python
import sys
import os
import random
import pickle
sys.path.append(os.path.abspath(os.path.join('../vggish_imports/')))
sys.path.append(os.path.abspath(os.path.join('../classification_models/')))
from vggish_similarity_model import findTopKSimilarSongs
from vggish_similarity_model import findTopKSimilarSongsWithinGenre
```

## General song similarity

We will show the results of out song similarity model across 1 randonly chosen song from each genre. The similar songs do not have to be within the same genre.

### Choose 1 random song from each genre

```python
genres = os.listdir('../Data/genres_original')

sampled_songs = {}

for genre in genres:
    songs = os.listdir('../Data/genres_original/' + genre)
    sampled_songs[genre] = random.sample(songs, 1)

print(sampled_songs)
```

```python
results = {}
for genre in genres:
    results[genre] = findTopKSimilarSongs(f'../Data/genres_original/{genre}/{sampled_songs[genre][0]}', "cosine_similarity", 10, print_results=False)
```

```python
for genre in genres:
    print(f"Genre: {genre}")
    for result in results[genre]:
        print(result)
    print("\n")
```

As we can see, songs within certain genres get similar songs only within its own genre whereas songs in other genres get a mix from different genres. Even though some songs' most similar songs are in different genres, they still sound similar, which shows how the model is able to indentify similar sounding songs. 

Note: the most similar song is the same song as the input songs since we are using songs from the same dataset that we are finding similar songs from.



### Similar songs from within the same genre

Now we will try to found the most similar songs from the same genre as the songs. This model involve first classifying the audio, then finding the most similar songs within the same genre.

We will use the same songs as above.

```python
results_2 = {}
for genre in genres:
    results_2[genre] = findTopKSimilarSongsWithinGenre(f'../Data/genres_original/{genre}/{sampled_songs[genre][0]}', "cosine_similarity", 10, print_results=False)
```

```python
for genre in genres:
    print(f"Genre: {genre}")
    for result in results_2[genre]:
        print(result)
    print("\n")
```

Now, this might not give use the most acoustically similar songs but it will give use the most similar songs in the same genre, which we believe will make it more likely for a person to like the songs (recommending songs from the same genre).

We can also see that out model was successful at identifying and classifying the genre of each of the songs. Obvsiously, classifying the genre correctly is very important for recommending relevant songs.
